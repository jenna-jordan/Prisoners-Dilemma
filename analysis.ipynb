```python
r1 = pd.read_csv('allStrategiesOnce.csv')
r2 = pd.read_csv('allStrategiesRandom.csv')
r3 = pd.read_csv('allStrategiesOnce_modeP.csv')
r4 = pd.read_csv('allStrategiesRandom_modeP.csv')
```

## Tournament Rankings when all strategies are played once, mode = misimplementation

```python
players1 = r1.groupby('PlayerStrategy').agg({'PlayerScore': 'mean', 'PlayerWinRate': 'mean', 'PlayerLossRate': 'mean'})
players1 = players1.sort_values(by='PlayerScore', ascending=False)
players1
```

## Tournament Rankings when all strategies are played 2-5 times, mode = misimplementation

```python
players2 = r2.groupby('PlayerStrategy').agg({'PlayerStrategy': 'count', 'PlayerScore': 'mean', 'PlayerWinRate': 'mean', 'PlayerLossRate': 'mean'})
players2 = players2.sort_values(by='PlayerScore', ascending=False)
players2
```
